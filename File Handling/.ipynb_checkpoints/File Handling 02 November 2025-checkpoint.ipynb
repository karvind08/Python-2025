```python
# Write a python program to combine each line from first file with corresponding line in the second file.
```

```python
f1 = open('data.txt','r')
f2 = open('file.txt','r')
f3 = open('merge.txt','w')
lines1 = f1.readlines()
lines2 = f2.readlines()
for i in range(len(lines1)):
    m=lines1[i]+lines2[i]
    f3.write(m)
f1.close()
f2.close()
f3.close()
```

```python
# Write a python program to read content from a text file and write those lines in other file that does not contain 'z' letter.
```

```python
f = open('data.txt','r')
f1 = open('nota.txt','w')
while True:
    line = f.readline()
    if line=='':
        break
    if 'n' not in line:
        f1.write(line)
f.close()
```
