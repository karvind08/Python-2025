```python
def readfile(file,mode):
    f = open(file,mode)
    data = f.read()
    print(data)
    
file = input('Enter file name with extension: ')
mode = input('Enter the mode')
readfile(file,mode)
```

```python
# Write a function in Python to read the content from a text file and print the number of vowels,consonants, upper and lowercase letters from file.
```

```python
def printletters():
    f = open('data.txt','r')
    data = f.read()
    v = 0
    c = 0
    l = 0
    u = 0
    for i in data:
        if i.isupper():
            u+=1
        elif i.islower():
            l+=1
        #i = i.lower()
        if i in 'aeiouAEIOU':
            v+=1
        elif i in 'bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ':
            c+=1
    print('The uppercase letters are: ',u)
    print('The lowercase letters are: ',l)
    print('The vowels are: ',v)
    print('The consonants are: ',c)
    
printletters()
```
