```python
import pickle
```

```python
def writedata():
    with open('db.dat','wb') as f:
        records = []
        while True:
            r = input()
            name = input()
            marks = input()
            data = [r,name,marks]
            records.append(data)
            pickle.dump(records,f)
            ch = input('Press y or n')
            if ch =='n':
                break

writedata()
            
```

```python
def readdata():
    with open('db.dat','rb') as f:
        rec = pickle.load(f)
        for i in rec:
            print(i)
readdata()
```

```python
def readdata():
    try:
        with open('db.dat','rb') as f:
            while True:
                rec = pickle.load(f)
                for i in rec:
                    print(i)
    except Exception:
        f.close()
readdata()
```
