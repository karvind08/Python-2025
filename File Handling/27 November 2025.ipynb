```python
import pickle
def writedata():
    with open('db1.dat','wb') as f:
        records = [] 
        while True:
            r = int(input())
            name = input()
            marks = input()
            data = [r,name,marks]
            records.append(data)
            ch = input('Press y and n')
            if ch=='n':
                break
            else:
                pickle.dump(records,f)

writedata()
```

```python
def readdata():
    with open('db1.dat','rb') as f:
        res = pickle.load(f)
        print(res)

readdata()
```
