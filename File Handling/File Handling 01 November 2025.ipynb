```python
# Write a function in Python to read the content from a text file and print the number of vowels,consonants, upper and lowercase letters from file.
```

```python
def printletters():
    f = open('data.txt','r')
    data = f.read()
    v = 0
    c = 0
    l = 0
    u = 0
    for i in data:
        if i.isupper():
            u+=1
        elif i.islower():
            l+=1
        i = i.lower()
        if i in 'aeiou':
            v+=1
        elif i in 'bcdfghjklmnpqrstvwxyz':
            c+=1
    print('The uppercase letters are: ',u)
    print('The lowercase letters are: ',l)
    print('The vowels are: ',v)
    print('The consonants are: ',c)
    f.close()
    
printletters()
```

```python
# Write a python function to read content from text file and calculate & display sum of all even numbers present in a text file.
```

```python
def funeven():
    f=open('file.txt','r')
    data=f.read()
    s=0
    for i in data:
        if i.isdigit() and int(i) % 2 == 0:
            s+=int(i)

    print('The sum is: ',s)
  
funeven()
```

```python
# Write a python program to replace all lowercase letters with their respective uppercase letters in the same file.
```

```python
f = open('file1.txt','r+')
data=f.read() 
# after reading the file, file pointer will be at the end of file. so to send it in the begining we will use seek()
f.seek(0)
for i in data:
    if i.islower():
        f.write(i.upper())
    else:
        f.write(i)
f.close()
```

```python
# Write a python function to read content from text file. all lowercase letters should be stored in lower.txt file whereas all uppercase letters should be stored in upper.txt file and other letters should be stored in others.txt
```

```python
f = open('file.txt','r')
fl = open('lower.txt','w')
fu = open('upper.txt','w')
fo = open('others.txt','w')
data = f.read()
for i in data:
    if i.islower():
        fl.write(i)
    elif i.isupper():
        fu.write(i)
    else:
        fo.write(i)
f.close()
fl.close()
fu.close()
fo.close()
```

```python
# Write a python program to display all records in a file along with line or record number.
```

```python
f = open('file.txt','r')
count = 1
while True:
    line = f.readline()
    if line =='':
        break
    else:
        print(count,':',line)
        count+=1
f.close()
```
