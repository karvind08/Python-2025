```python
from matplotlib import pyplot as plt
```

```python
x1 = [2,6,9,13]
y1 = [14,50,67,87]
x2 = [5,19,12,21]
y2 = [21,45,71,89]
plt.bar(x1,y1,color='g',label='First Bar')
plt.bar(x2,y2,color='c',label='Second Bar')
plt.title('Bar Plot')
plt.legend()
plt.xlabel('X-Axis')
plt.xlabel('Y-Axis')
plt.show()
```

```python
students = ['A','B','C','D']
marks = [67,78,98,99]
plt.bar(students,marks,color='r',label='First Bar')
plt.title('Bar Plot')
plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
students = ['A','B','C','D']
marks = [67,78,98,99]
plt.barh(students,marks,color='r',label='First Bar')
plt.title('Bar Plot')
plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# Pie Plot
```

```python
slices = [10,21,30,40,50]
names = ['A','B','C','D','E']
colrs = ['r','g','b','y','c']
plt.pie(slices,
        labels = names,
        colors = colrs,
        startangle = 60,
        shadow = True,
        explode = (0,0,0.1,0,0),
        autopct = '%.2f%%'
       )
plt.show()
```

```python
slices = [10,21,30,40,50]
names = ['A','B','C','D','E']
colrs = ['r','g','b','y','c']
plt.pie(slices,
        labels = names,
        colors = colrs,
        shadow = True,
        startangle = 90,
        explode = (0,0,0,0,0.2),
        autopct = '%.2f%%',
       )
plt.show()
```
