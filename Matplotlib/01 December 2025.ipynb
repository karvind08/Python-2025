```python
from matplotlib import pyplot as plt
```

```python
x1 = [2,6,9,13]
y1 = [14,50,67,87]
x2 = [5,19,12,21]
y2 = [21,45,71,89]
plt.bar(x1,y1,color='g',label='First Bar')
plt.bar(x2,y2,color='c',label='Second Bar')
plt.title('Bar Plot')
plt.legend()
plt.xlabel('X-Axis')
plt.xlabel('Y-Axis')
plt.show()
```

```python
students = ['A','B','C','D']
marks = [67,78,98,99]
plt.bar(students,marks,color='r',label='First Bar')
plt.title('Bar Plot')
plt.legend()
plt.xlabel('X-Axis')
plt.xlabel('Y-Axis')
plt.show()
```
