```python
from matplotlib import style
style.use('grayscale')
x1 = [1,6,13,18]
y1 = [10,12,56,76]
x2 = [100,200,45,6]
y2 = [4,56,98,87]
plt.plot(x1,y1,'g',label='Line 1',linewidth=3)
plt.plot(x2,y2,'b',label='Line 2',linewidth=3)
plt.legend()
plt.grid(True,color='r')
plt.show()
```

```python
from matplotlib import style
style.use('bmh')
x1 = [1,6,13,18]
y1 = [10,12,56,76]
x2 = [100,200,45,6]
y2 = [4,56,98,87]
plt.bar(x1,y1,label='One')
plt.bar(x2,y2,label='Two')
plt.legend()
plt.title('bar Plot')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x1 = [3,6,13,18]
y1 = [10,12,56,76]
x2 = [100,200,45,6]
y2 = [4,56,98,87]
plt.bar(x1,y1,label='One')
plt.bar(x2,y2,label='Two')
plt.legend()
plt.title('bar Plot')
plt.show()
```
