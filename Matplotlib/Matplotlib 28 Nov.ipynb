```python
from matplotlib import pyplot as plt
```

```python
x = [1,5,9]
y = [10,-56,98]
plt.plot(x,y)
plt.show()
```

```python
x = [1,5,9]
y = [10,-56,98]
plt.plot(x,y)
plt.title('Plot 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
x = [1,5,9]
y = [10,-56,98]
plt.plot(x,y,label='Label')
plt.title('Plot 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x = [1,6,13,18]
y = [10,12,56,76]
plt.plot(x,y,label='Line 1','g',linewidth=6)
plt.show()
```
